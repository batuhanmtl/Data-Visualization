For the graphs we will draw, we will first need the pyplot subpackage in the matplotlib package. First we need to import this package.
For convenience it is usually exported as plt for the matplotlib.pyplot package.

In the example below, let's draw a line graph of Turkey's economic growth rates between 1999 and 2016.

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.arange(1999,2017)

growth = [-3.4,6.6,-6,6.4,5.6,9.6,9,7.1,5,0.8,-4.7,8.5,11.1,4.8,8.5,5.2,6.1,3.2]

plt.plot(years,growth)
plt.xlabel("Years")
plt.ylabel("Growth")
plt.title("Turkey's Growth Rates (1999-2016)")
plt.show()
```

You may have noticed that the years are oddly written as 2002.5,2007.5 in the chart above. The .xticks() and .yticks() methods can be used to fix this.

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.arange(1999,2017)

growth = [-3.4,6.6,-6,6.4,5.6,9.6,9,7.1,5,0.8,-4.7,8.5,11.1,4.8,8.5,5.2,6.1,3.2]

plt.plot(years,growth)
plt.xlabel("Years")
plt.ylabel("Growth")
plt.title("Turkey's Growth Rates (1999-2016)")
plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016])
plt.show()
```

Now let's draw the two graphs together.

```python
import numpy as np
import matplotlib.pyplot as plt

y1 = np.array([25.7,23.8,22.4,25.6,26.1,27.3,29.9,30.1,
               31.2,31.8,32.2,34.9,33.4,35.4,33.8,36.2,
               37.3,39.8,40.4,40.2,40.3,41.3,44.5,47.3,
               47.9,45.2,44.7,46.4,50.5,52.5,56.9,56.4,
               55.2,54.8,54.2,52.9,51.9,52.4,52.5,55.6])
y2 = np.array([24.7,25.4,24.3,25.5,27.8,29.7,32.3,35.4,
               38,40.4,42.1,44.4,46.6,47,46.3,45.3,
               44.2,42.6,40.2,39.3,39.2,38.2,38.9,38.1,
               38.2,37.8,37,36.2,37.4,38.8,37.4,37.3,
               37,35.4,32.5,30.4,28.2,27.3,27.4,28.4])
years = np.arange(1978,2018)
plt.plot(years,y1,'r') # r = red
plt.plot(years,y2,'b') # b = blue
plt.show()
```

It is possible to show the 2 graphics in the figure side by side in 2 different frames. We will use the plt.axes() method for this.
In plt.axes() method, it is necessary to introduce 4 arguments in the list. x,y,width,height.
The first 2 arguments, x and y, specify the position of the bottom left corner of the chart relative to the main chart.
The width and height also represent the ratio of the width and height of the sub chart to the width and height of the main chart.

```python
plt.axes([0.05,0.05,0.4,0.9])
plt.plot(years,y1,'r')
plt.axes([0.55,0.05,0.4,0.9])
plt.plot(years,y2,'b')
plt.show()
```

With the Axes method, it is also possible to draw another graphic in a smaller size inside a graphic. This allows creating beautiful visuals.
Notice that we set the background color of the graphic with the facecolor = 'y' argument in the code below.

```python
plt.axes([0.05,0.05,0.9,0.9])
plt.plot(years,y1,'r')
plt.axes([0.1,0.55,0.3,0.3],facecolor = 'y')
plt.plot(years,y2,'b')
plt.show()
```

With the code below, we first specify the label we want to appear on the graph with the label argument for each data graph.
We use the plt.legend() command to take place in the graph. To specify the location of the label, we set the loc argument as upper left.

```python
plt.plot(years,y1,'r--',label='ABCD')
plt.plot(years,y2,'g:',label='xyzt')
plt.legend(loc='upper left')
plt.show()
```

It is also possible to write text on the point we want in the graphics. This is a useful feature when we want to add annotations on the graphic.
The plt.annotate() method is used to add text on the graphic. The text to be added in the method and the xy argument and the x and y coordinates where the text will be placed are transmitted. This time, instead of adding legends as above, let's write their names as text on the relevant graphics.

```python
plt.plot(years,y1,'r--')
plt.plot(years,y2,'g:')
plt.annotate('ABCD',xy = (2010,50))
plt.annotate('xyzt',xy=(2010,38))
plt.show()
```

It is also possible to draw arrows with the text. For this, the coordinates of the text with the xytext argument,
the coordinates of the arrowhead with the xy argument, and the properties of the arrow with the arrowprops argument are transmitted as a dictionary.
The arrowprops argument also makes the arrow appear. Therefore, we need to write this argument even if it is empty.

```python
plt.plot(years,y1,'r--')
plt.plot(years,y2,'b:')
plt.annotate('ABCD',xytext=(1995,50),xy=(2000,45),arrowprops={'color':'red'})
plt.annotate('xyzt',xytext=(2000,30),xy=(2005,35),arrowprops = {'color':'blue'})
plt.show()
```

We can draw various graphs using the data of Tesla stock. For example, let's draw daily closing information.

```python
import pandas as pd
import matplotlib.pyplot as plt
tesla = pd.read_csv('tesla.csv')
plt.plot(tesla['Close'])
plt.show()
```

Now let's add the title and y-axis information to the chart.

```python
tesla = pd.read_csv('tesla.csv',index_col='Date',parse_dates = True)
tesla['Close'].plot(title = 'Tesla Closing Price')
plt.ylabel('Closing Price')
plt.xlabel('Date')
plt.show()
```
