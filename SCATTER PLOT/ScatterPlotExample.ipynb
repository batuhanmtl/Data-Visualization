In the line graph, the data points are combined with the line segments. Instead, the graphs that show each data point individually in shit or any other way we want are called point graphs or scatter plots. The economic growth data of Turkey between 1999-2016 in the previous section Let's draw it as a scatter plot. For the scatter plot, it will be enough to use the .scatter() method instead of the .plot() method.

```python
import matplotlib.pyplot as plt

years = range(1999,2017)

growth = [-3.4,6.6,-6,6.4,5.6,9.6,9,7.1,5,0.8,-4.7,8.5,11.1,4.8,8.5,5.2,6.1,3.2]

plt.scatter(years,growth,color='green')
plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016])
plt.yticks([-5.0,0.0,5.0,10.0],['-%5','%0','%5','%10'])
plt.xlabel("Years")
plt.ylabel("Growth Rate")
plt.title("Turkey's Growth Rates (1999-2016)")
plt.show()
```

We can use .scatter() and .plot() method together to connect points with lines.

```python
plt.scatter(years,growth,color='green')
plt.plot(years,growth)
plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016])
plt.yticks([-5.0,0.0,5.0,10.0],['-%5','%0','%5','%10'])
plt.xlabel("Years")
plt.ylabel("Growth Rate")
plt.title("Turkey's Growth Rates (1999-2016)")
plt.show()
```
