The .meshgrid() method is used to calculate a two-variable function in Python. Now let's say we want to plot the following function consisting of x and y variables.


$$Z =\cos \left( \frac{\sqrt{X^2+Y^2}}{\pi} \right )$$


Let X and Y values be between -10 and 10. First, we generate X and Y values with the np.linspace() method.

```python
import numpy as np
import matplotlib.pyplot as plt

u = np.linspace(-10,10,100)
v = np.linspace(-10,10,100)
```

Now we will create (X,Y) pairs by creating the number sequences u and v. We will use the np.meshgrid() function for this. So we can use the X and Y variables in the above function.

```python
X, Y = np.meshgrid(u,v)
```

Let's calculate the Z variable by using the X and Y variables in the above equation. Pay attention to the Numpy functions we use. Let's use the np.hypot(X,Y) function instead of the np.sqrt(X**2 + Y**2) function in the equation.

```python
z = np.cos(np.hypot(X,Y)/np.pi)
```

Now let's draw the level graph. We can use plt.pcolor() or plt.imshow() functions for this. Below are graphs drawn with both. Dark colors show low values in pixels and light colors show high values. You can do this with plt.colorbar() You can also tell from the color scale we created.

```python
plt.pcolor(z)
plt.colorbar()
plt.xlabel('Level Chart')
plt.show()
```

Now let's plot the graph with plt.imshow().

```python
plt.imshow(z)
plt.xlabel('Level Chart')
plt.show()
```
