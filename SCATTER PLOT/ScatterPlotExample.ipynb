In the line graph, the data points are combined with the line segments. Instead, the graphs that show each data point individually in shit or any other way we want are called point graphs or scatter plots. The economic growth data of Turkey between 1999-2016 in the previous section Let's draw it as a scatter plot. For the scatter plot, it will be enough to use the .scatter() method instead of the .plot() method.

```python
import matplotlib.pyplot as plt

years = range(1999,2017)

growth = [-3.4,6.6,-6,6.4,5.6,9.6,9,7.1,5,0.8,-4.7,8.5,11.1,4.8,8.5,5.2,6.1,3.2]

plt.scatter(years,growth,color='green')
plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016])
plt.yticks([-5.0,0.0,5.0,10.0],['-%5','%0','%5','%10'])
plt.xlabel("Years")
plt.ylabel("Growth Rate")
plt.title("Turkey's Growth Rates (1999-2016)")
plt.show()
```

We can use .scatter() and .plot() method together to connect points with lines.

```python
plt.scatter(years,growth,color='green')
plt.plot(years,growth)
plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016])
plt.yticks([-5.0,0.0,5.0,10.0],['-%5','%0','%5','%10'])
plt.xlabel("Years")
plt.ylabel("Growth Rate")
plt.title("Turkey's Growth Rates (1999-2016)")
plt.show()
```

From time to time, we want to show two properties on the x and y axes in a scatterplot, while we want to show the third property as color or size. In the example we will use for this below, we will use the Boston house prices dataset from the sample datasets in the sklearn package. While showing the average price on the x-axis and the average price on the y-axis, we will reflect the crime rate in the city as size.
We will use the sklearn.datasets package to get the dataset, the pandas package to make it available for plotting, and the matplotlib.pyplot packages for plotting.

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston

boston_housing = load_boston()
boston = pd.DataFrame(boston_housing.data)
boston.columns = boston_housing.feature_names
boston['Price'] = boston_housing.target

plt.scatter(boston.RM,boston.Price,s =boston.CRIM,color='red')
plt.xlabel('Average Number Of Rooms')
plt.ylabel('Price')
plt.show()
```

As seen in the graphic above, the crime rates in each city are reflected in the size of the figure. We do this with the s, that is, 'size' argument in the plt.scatter() function.

Now let's look at an example where we show different species with different colors. In the example below, the Petal Width and Petal Length features in the iris dataset, which is widely used in artificial learning, are drawn on the x and y axes, and different types of flowers are shown in different colors. A different color will be used for each different species in the column.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
iris = pd.DataFrame(iris_data.data)
iris.columns = ['sepal_length','sepal_width','petal_length','petal_width']
iris['species'] = iris_data.target

plt.scatter(iris.petal_length,iris.petal_width,c=iris.species)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Data Set')
plt.show()
```

From the graph, it can be seen that there is a high and positive correlation between petal length and petal width. It is also seen that iris species in different groups are clustered in different areas. As a result, the type of this flower can be predicted with the width and height information.

Let's plot the groups in the iris data set in a slightly different way this time with the subplot method.

```python
#Let's split the chart into sub-charts consisting of 1 row and 3 columns.
figure,axis = plt.subplots(nrows=1,ncols=3)

#Let's make a selection for each point and apply the .plot method.

iris[iris['species']=='Iris-setosa'].plot(ax = axis[0],x = 'sepal_length',y ='sepal_width',kind='scatter',color='b')
iris[iris['species']=='Iris-versicolor'].plot(ax = axis[1],x = 'sepal_length',y ='sepal_width',kind='scatter',color='r')
iris[iris['species']=='Iris-virginica'].plot(ax = axis[2],x = 'sepal_length',y ='sepal_width',kind='scatter',color='g')
plt.show()
```

With the style.use() command in the matplotlib module, we can make the graphic plotted in different predefined themes. Now let's draw the iris graphic that we drew earlier in different themes (ggplot theme).

```python
plt.scatter(iris.iloc[0:49,2],iris.iloc[0:49,3],label='Setosa',c='r')
plt.scatter(iris.iloc[50:99,2],iris.iloc[50:99,3],label='Versicolor',c='y')
plt.scatter(iris.iloc[100:149,2],iris.iloc[100:149,3],label='Virginica',c='g')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Data')
plt.legend(loc='upper left')
plt.style.use('ggplot')
plt.show()
```
